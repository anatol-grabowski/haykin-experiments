```python
import numpy as np
import pandas as pd
import scipy.special
```

```python
def covers_prob(N, m1):
    a = (1/2)**(N-1)
    b = 0
    for m in range(m1-1 +1):
#         print(m)
        b += scipy.special.binom(N-1, m)
    return a*b

Ns = np.arange(1, 40)
m1s = np.arange(1, 20)

df = pd.DataFrame(index=Ns, columns=m1s)
df.index.name = 'N'
df.columns = pd.MultiIndex.from_product([df.columns], names=['m1'])
for N in Ns:
    for m1 in m1s:
        df.loc[N, m1] = covers_prob(N, m1)
        
df
```
